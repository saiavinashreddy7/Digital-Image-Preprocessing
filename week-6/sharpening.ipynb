```python
#Write programs to perform following sharpening operations on a grayscale image
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r"D:\Avinash\Digital-Image-Preprocessing\assets\one.jpg", cv2.IMREAD_GRAYSCALE)
#Laplacian filter
laplacian_filtered = cv2.Laplacian(image, cv2.CV_64F)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_filtered = cv2.magnitude(sobel_x, sobel_y)
#Filtering using first order derivative operators such as sobel and prewitt mask.
prewitt_x = cv2.filter2D(image, -1, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
prewitt_y = cv2.filter2D(image, -1, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))
prewitt_filtered = cv2.magnitude(prewitt_x.astype(np.float64), prewitt_y.astype(np.float64))

plt.figure(figsize=(10, 7))
plt.subplot(2, 3, 1), plt.imshow(image, cmap='gray'), plt.title("Original Image")
plt.subplot(2, 3, 2), plt.imshow(laplacian_filtered, cmap='gray'), plt.title("Laplacian Filter")
plt.subplot(2, 3, 3), plt.imshow(sobel_filtered, cmap='gray'), plt.title("Sobel Filter")
plt.subplot(2, 3, 4), plt.imshow(prewitt_filtered, cmap='gray'), plt.title("Prewitt Filter")
plt.tight_layout()
plt.show()
```
