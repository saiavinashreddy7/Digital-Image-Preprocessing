```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image_path = "image.png"  
image = cv2.imread(r"D:\Avinash\Digital-Image-Preprocessing\assets\grayscale.jpg")
```

```python
min_val, max_val = np.min(image), np.max(image)
contrast_stretch = ((image - min_val) / (max_val - min_val) * 255).astype(np.uint8)
cv2.imwrite("contrast_stretch.png", contrast_stretch)
plt.imshow(contrast_stretch, cmap='gray')
plt.title("Contrast Stretching")
plt.axis('off')
plt.show()
```
