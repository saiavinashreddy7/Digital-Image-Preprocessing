```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image_path = "image.png"  
image = cv2.imread(r"D:\Avinash\Digital-Image-Preprocessing\assets\grayscale.jpg")
```

```python
c_log = 45
inv_log_image = np.exp(image.astype(np.float32) / c_log) - 1
inv_log_image = cv2.normalize(inv_log_image, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
cv2.imwrite("inv_log_image.png", inv_log_image)
plt.imshow(inv_log_image, cmap='gray')
plt.title("Inverse Log Transformation")
plt.axis('off')
plt.show()
```
