```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image_path = "image.png"  
image = cv2.imread(r"D:\Avinash\Digital-Image-Preprocessing\assets\grayscale.jpg")
```

```python
gamma_1, gamma_2 = 0.5, 2.0
power_law_1 = np.power(image.astype(np.float32) / 255, gamma_1) * 255
power_law_1 = power_law_1.astype(np.uint8)
cv2.imwrite("power_law_1.png", power_law_1)
plt.imshow(power_law_1, cmap='gray')
plt.title("Power Law (Gamma=0.5)")
plt.axis('off')
plt.show()
power_law_2 = np.power(image.astype(np.float32) / 255, gamma_2) * 255
power_law_2 = power_law_2.astype(np.uint8)
cv2.imwrite("power_law_2.png", power_law_2)
plt.imshow(power_law_2, cmap='gray')
plt.title("Power Law (Gamma=2.0)")
plt.axis('off')
plt.show()
```
