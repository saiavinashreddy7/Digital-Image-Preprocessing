```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image_path = "image.png"  
image = cv2.imread(r"D:\Avinash\Digital-Image-Preprocessing\assets\grayscale.jpg")
if image is None:
    raise ValueError("Error loading image. Check the file path.")
```

```python
c_log = 45
log_image = c_log * np.log1p(image.astype(np.float32))
log_image = cv2.normalize(log_image, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
cv2.imwrite("log_image.png", log_image)
plt.imshow(log_image, cmap='gray')
plt.title("Log Transformation")
plt.axis('off')
plt.show()
```
