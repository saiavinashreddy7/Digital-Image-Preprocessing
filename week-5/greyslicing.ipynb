```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image_path = "image.png"  
image = cv2.imread(r"D:\Avinash\Digital-Image-Preprocessing\assets\grayscale.jpg")
```

```python
gray_sliced = np.where((image > 50) & (image < 150), 255, image)
cv2.imwrite("gray_sliced.png", gray_sliced)
plt.imshow(gray_sliced, cmap='gray')
plt.title("Gray Level Slicing")
plt.axis('off')
plt.show()
```
