```python
import numpy as np
import matplotlib.pyplot as plt
def create_image_with_box(image_size=1000, box_size=100):
    image = np.zeros((image_size, image_size))
    start_row = (image_size - box_size) // 2
    start_col = (image_size - box_size) // 2
    image[start_row:start_row+box_size, start_col:start_col+box_size] = 1
    return image
image = create_image_with_box()
plt.figure(figsize=(6, 6))
plt.imshow(image, cmap='gray')
plt.title("Original Image with White Box")
plt.axis('off')
plt.show()

 
```
