```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
image = cv2.imread(r"D:\Avinash\Digital-Image-Preprocessing\assets\one.jpg")
if image is None:
    print("Error: Could not load the image. Check the file path.")
else:
    blue_plane, green_plane, red_plane = cv2.split(image)
    blue_color = np.zeros_like(image)
    blue_color[:, :, 0] = blue_plane  
    green_color = np.zeros_like(image)
    green_color[:, :, 1] = green_plane  
    red_color = np.zeros_like(image)
    red_color[:, :, 2] = red_plane  

     
    plt.figure(figsize=(15, 5))

 
    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(blue_color, cv2.COLOR_BGR2RGB))
    plt.title("Blue Plane")
    plt.axis('off')

    
    plt.subplot(1, 3, 2)
    plt.imshow(cv2.cvtColor(green_color, cv2.COLOR_BGR2RGB))
    plt.title("Green Plane")
    plt.axis('off')

     
    plt.subplot(1, 3, 3)
    plt.imshow(cv2.cvtColor(red_color, cv2.COLOR_BGR2RGB))
    plt.title("Red Plane")
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
