```python
#Create a white image of size 256x256, with black box of size 58x58 at centre.
import numpy as np
import matplotlib.pyplot as plt

image_size = 256
box_size = 58
image = np.ones((image_size, image_size))

start_row = (image_size - box_size) // 2
start_col = (image_size - box_size) // 2

image[start_row:start_row+box_size, start_col:start_col+box_size] = 0

plt.figure(figsize=(6, 6))
plt.imshow(image, cmap='gray')
plt.title("White Image with Centered Black Box")
plt.axis('off')
plt.show()
```
