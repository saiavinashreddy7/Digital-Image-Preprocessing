```python
#Subtraction of A and B
import numpy as np
import matplotlib.pyplot as plt

image_size = 1024
line_width = 128
image_a = np.zeros((image_size, image_size))
for i in range(0, image_size, 2 * line_width):
    image_a[i:i+line_width, :] = 1
image_b = np.zeros((image_size, image_size))
for j in range(0, image_size, 2 * line_width):
    image_b[:, j:j+line_width] = 1
image_subtraction = np.clip(image_a - image_b, 0, 1)  # Ensures pixel values are between [0, 1]
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(image_a, cmap='gray')
axes[0].set_title("Image A (Horizontal Stripes)")
axes[0].axis('off')
axes[1].imshow(image_b, cmap='gray')
axes[1].set_title("Image B (Vertical Stripes)")
axes[1].axis('off')
axes[2].imshow(image_subtraction, cmap='gray')
axes[2].set_title("Image Subtraction (A - B)")
axes[2].axis('off')
plt.tight_layout()
plt.show()
```
